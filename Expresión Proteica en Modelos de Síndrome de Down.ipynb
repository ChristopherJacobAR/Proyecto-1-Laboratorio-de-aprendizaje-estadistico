## 1. El Síndrome de Down: Perspectiva Biológica, Química y Neurocientífica

El síndrome de Down es una condición genética causada por la trisomía del cromosoma 21, lo que conduce a alteraciones en el desarrollo físico y cognitivo. Las alteraciones en la estructura cromosómica afectan múltiples procesos celulares y moleculares, lo que se refleja en cambios en la expresión proteica y en la funcionalidad de diversas vías de señalización.

### Aspectos Biológicos y Neurocientíficos

- **Desarrollo Cerebral y Conectividad Neuronal:**  
  La presencia de una copia extra del cromosoma 21 conduce a la sobreexpresión de ciertos genes, lo que afecta la formación y la conectividad de las redes neuronales. Esto puede traducirse en deficiencias en la memoria, el aprendizaje y otras funciones cognitivas.
  
- **Síntesis y Expresión Proteica:**  
  Se observa un patrón alterado de expresión proteica en áreas críticas del cerebro, lo que puede influir en procesos como la sinapsis y la plasticidad neuronal. La cuantificación de estas proteínas, mediante técnicas de expresión proteica, es crucial para identificar biomarcadores y entender los mecanismos subyacentes del síndrome de Down.

### Aspectos Físico-Químicos

- **Modificaciones Postraduccionales:**  
  A nivel molecular, las proteínas no solo se sintetizan, sino que también sufren modificaciones postraduccionales (como fosforilación, acetilación, ubiquitinación, etc.). Estas modificaciones alteran la conformación, estabilidad y función de las proteínas.  
  - **Ejemplo:** La fosforilación se puede representar mediante la ecuación:  
    $$
    \text{Proteína} + \text{ATP} \xrightarrow{\text{Quinasas}} \text{Proteína-Fosforilada} + \text{ADP}
    $$
  
- **Procesos de Señalización Celular y Reacciones Enzimáticas:**  
  Los procesos celulares se rigen por leyes físicas y químicas. Por ejemplo, las reacciones enzimáticas se modelan a menudo usando la ecuación de Michaelis-Menten, que describe la cinética de la conversión de sustrato en producto:
  
  $$
  v = \frac{V_{max} \cdot [S]}{K_m + [S]}
  $$
  
  Donde:
  - $v$ es la velocidad de la reacción.
  - $V_{max}$ es la velocidad máxima.
  - $[S]$ es la concentración del sustrato.
  - $K_m$ es la constante de Michaelis-Menten (la concentración de sustrato a la cual la velocidad es la mitad de $V_{max}$).

- **Dinámica y Equilibrio Químico:**  
  Las reacciones bioquímicas se rigen por principios termodinámicos. El equilibrio químico se describe mediante la constante de equilibrio $K$, definida por:
  
  $$
  K = \frac{[C]^c [D]^d}{[A]^a [B]^b}
  $$
  
  donde $a$, $b$, $c$ y $d$ son los coeficientes estequiométricos de la reacción:
  
  $$
  aA + bB \rightleftharpoons cC + dD
  $$

### Relación del Síndrome de Down con la Investigación

En el contexto del síndrome de Down, estas alteraciones bioquímicas y moleculares pueden llevar a:
- Cambios en las vías de señalización que afectan el desarrollo neuronal.
- Disminución en la eficacia de los procesos enzimáticos y en la dinámica de la síntesis y degradación de proteínas.
- Patrones anómalos en la modificación postraduccional de proteínas clave, lo que repercute en la estructura y función de las células cerebrales.

## Representación Interactiva de Procesos Químicos y Reacciones

A continuación se muestra un ejemplo en Python que utiliza Plotly e ipywidgets para crear gráficos interactivos que explican, por ejemplo, la cinética enzimática basada en la ecuación de Michaelis-Menten.



```python
# Importar las bibliotecas necesarias
import numpy as np
import plotly.graph_objs as go
import ipywidgets as widgets
from IPython.display import display

def michaelis_menten(S, Vmax, Km):
    return (Vmax * S) / (Km + S)

def update_plot(Vmax=1.0, Km=0.5):
    S = np.linspace(0, 10, 200)  # Concentración del sustrato
    v = michaelis_menten(S, Vmax, Km)
    
    trace = go.Scatter(x=S, y=v, mode='lines', name='Velocidad de reacción')
    layout = go.Layout(
        title=f'Cinética enzimática: Michaelis-Menten<br>Vmax={Vmax:.2f}, Km={Km:.2f}',
        xaxis=dict(title='Concentración de Sustrato [S]'),
        yaxis=dict(title='Velocidad de reacción (v)')
    )
    fig = go.Figure(data=[trace], layout=layout)
    fig.show()

Vmax_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Vmax:')
Km_slider = widgets.FloatSlider(value=0.5, min=0.1, max=5.0, step=0.1, description='Km:')

ui = widgets.HBox([Vmax_slider, Km_slider])
out = widgets.interactive_output(update_plot, {'Vmax': Vmax_slider, 'Km': Km_slider})

display(ui, out)
```

## Explicación de la Gráfica: Cinética Enzimática de Michaelis-Menten

La gráfica muestra la relación entre la **velocidad de reacción** $(v)$ y la **concentración de sustrato** $([S])$ en un sistema enzimático que sigue la cinética de Michaelis-Menten. En el eje **horizontal** se representa $[S]$, mientras que en el eje **vertical** se observa la **velocidad de la reacción** $(v)$.

### Ecuación de Michaelis-Menten

La cinética de Michaelis-Menten se describe mediante la siguiente fórmula:

$$
v = \frac{V_{max} \cdot [S]}{K_m + [S]}
$$

donde:

- $v$ es la velocidad de la reacción (velocidad a la que se forma el producto).
- $V_{max}$ es la velocidad máxima que se alcanza cuando la enzima está saturada de sustrato.
- $K_m$ (constante de Michaelis) es la concentración de sustrato a la cual la velocidad de reacción es la mitad de $V_{max}$.
- $[S]$ es la concentración de sustrato.

### Forma de la Curva

1. **Fase Inicial (Baja Concentración de Sustrato):**  
   - Cuando $[S]$ es bajo, la velocidad $v$ aumenta casi de forma lineal con $[S]$.  
   - La enzima no está saturada, por lo que cada incremento en la concentración de sustrato produce un incremento significativo en la velocidad.

2. **Zona de Saturación (Alta Concentración de Sustrato):**  
   - A medida que $[S]$ aumenta, la curva se acerca asintóticamente a $V_{max}$.  
   - En este punto, la mayoría de los sitios activos de la enzima están ocupados, por lo que un aumento adicional en $[S]$ no produce un incremento sustancial en la velocidad.

3. **Significado de $V_{max}$ y $K_m$:**  
   - **$V_{max}$**: Indica la máxima capacidad de la enzima para catalizar la reacción. Refleja la situación en la que todas las enzimas están saturadas con sustrato.  
   - **$K_m$**: Proporciona información sobre la afinidad de la enzima por el sustrato. Un valor de $K_m$ bajo implica alta afinidad (se necesita poca concentración de sustrato para alcanzar la mitad de $V_{max}$).

### Interpretación

- La **curva hiperbólica** evidencia cómo la velocidad de la reacción se incrementa rápidamente al principio, pero luego se **acerca a un límite** (no puede superar $V_{max}$).  
- Si el **$K_m$** es relativamente **bajo**, la enzima trabaja con alta eficiencia incluso a concentraciones moderadas de sustrato.  
- Si el **$K_m$** es **alto**, se necesita más sustrato para alcanzar la mitad de la velocidad máxima, indicando menor afinidad de la enzima por el sustrato.

En conjunto, la gráfica y la ecuación de Michaelis-Menten son fundamentales para comprender los **mecanismos de catálisis enzimática**, el **diseño de fármacos** y la **regulación de rutas metabólicas** en distintos contextos, incluido el estudio de alteraciones enzimáticas y proteicas en el síndrome de Down.




## Modelos Matemáticos y Estadísticos en el Análisis

### Regresión Lineal

La regresión lineal se utiliza para modelar la relación lineal entre la variable dependiente $Y$ (nivel de expresión proteica) y las variables independientes $X_i$ (factores experimentales). El modelo se expresa como:

$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \cdots + \beta_pX_p + \epsilon
$$

- **$\beta_0$:** Intersección (valor de $Y$ cuando todas las $X_i=0$).  
- **$\beta_i$:** Coeficiente del efecto de la variable $X_i$.  
- **$\epsilon$:** Término de error.

Este modelo permite evaluar el impacto individual de cada factor (por ejemplo, genotipo, tratamiento y comportamiento) sobre la expresión proteica.

---

### Regresión Polinomial

Cuando la relación entre $Y$ y $X$ presenta curvaturas o comportamientos no lineales, se utiliza la regresión polinomial. El modelo se extiende de la siguiente forma:

$$
Y = \beta_0 + \beta_1X + \beta_2X^2 + \cdots + \beta_nX^n + \epsilon
$$

Esta formulación permite capturar relaciones complejas, como aquellas derivadas de procesos bioquímicos y reacciones enzimáticas que no varían de forma lineal.

---

### Interacción de Factores

La interacción se modela cuando el efecto de una variable sobre $Y$ depende del valor de otra variable. Se incluye un término de interacción en el modelo:

$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \beta_3(X_1 \times X_2) + \epsilon
$$

Por ejemplo, $X_1$ puede representar el genotipo y $X_2$ el tratamiento; el término $X_1 \times X_2$ permite evaluar si el efecto del tratamiento varía según el genotipo del ratón.

---

### Significancia de Factores

Para determinar qué variables son determinantes, se utilizan pruebas estadísticas:
- **Prueba t:** Evalúa si cada coeficiente $\beta_i$ es significativamente diferente de cero.
- **Prueba F:** Analiza la significancia global del modelo.
- **Intervalos de confianza y p-valores:** Proporcionan evidencia de la confiabilidad de las estimaciones.

Estos métodos permiten validar la relevancia de cada factor y sus interacciones en la explicación de la variabilidad en la expresión proteica.

---

###  Regularización: Ridge, Lasso y ElasticNet

Cuando se manejan múltiples variables y se presentan problemas de multicolinealidad, la regularización ayuda a mejorar la generalización del modelo.

#### Ridge (Regresión de Cresta)

Se minimiza la siguiente función de costo:

$$
\min_{\beta} \left\{ \|Y - X\beta\|_2^2 + \lambda \|\beta\|_2^2 \right\}
$$

- **$\lambda$:** Parámetro de regularización que controla la penalización.

#### Lasso

Se minimiza:

$$
\min_{\beta} \left\{ \|Y - X\beta\|_2^2 + \lambda \|\beta\|_1 \right\}
$$

- La penalización $L1$ puede llevar algunos coeficientes a cero, facilitando la selección de variables.

#### ElasticNet

Combina ambas penalizaciones:

$$
\min_{\beta} \left\{ \|Y - X\beta\|_2^2 + \lambda_1 \|\beta\|_1 + \lambda_2 \|\beta\|_2^2 \right\}
$$

- Ofrece un balance entre la selección de variables (Lasso) y la estabilidad (Ridge).

Estos métodos son fundamentales para evitar el sobreajuste y garantizar que los modelos sean robustos y generalizables, especialmente en contextos con alta dimensionalidad y correlación entre variables.

# Modelado de la Expresión Proteica: Ajuste y Comparación de Modelos

Este análisis utiliza el dataset "Mice Protein Expression" para predecir la variable `NR1_N` (nivel de expresión de la subunidad NR1 del receptor NMDA) utilizando todas las variables numéricas disponibles (excepto `NR1_N` misma) como predictores. Se implementan tres tipos de modelos:
1. Regresión lineal (grado 1)
2. Regresión polinómica de grado 2
3. Regresión polinómica de grado 3

Para cada tipo se evalúan cuatro versiones:
- Sin penalización (OLS)
- Con Ridge
- Con Lasso
- Con ElasticNet

Se utiliza validación cruzada con K-Folds para comparar el desempeño de los modelos en términos de MSE y R². Además, se realiza un análisis de la significancia de factores (p-values, intervalos de confianza) usando `statsmodels` para el modelo sin penalización (OLS).

---




```python
numeric_cols = df.select_dtypes(include=['float64']).columns.tolist()
predictors = [col for col in numeric_cols if col != 'NR1_N']
target = 'NR1_N'

X = df[predictors]
y = df[target]

# --- Paso 2: Definir funciones para crear pipeline y evaluar modelos ---
def create_pipeline(estimator, poly_degree=1):
    steps = []
    if poly_degree > 1:
        steps.append(('poly', PolynomialFeatures(degree=poly_degree, include_bias=False)))
    steps.append(('scaler', StandardScaler()))
    steps.append(('model', estimator))
    return Pipeline(steps)

def evaluate_model(model, X, y, cv=5):
    kf = KFold(n_splits=cv, shuffle=True, random_state=42)
    mse_scores = -cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error')
    r2_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
    return mse_scores.mean(), r2_scores.mean()

# --- Paso 3: Ajustar múltiples modelos y recopilar resultados ---
model_versions = {
    'No_Penalization': LinearRegression(),
    'Ridge': Ridge(max_iter=10000),
    'Lasso': Lasso(max_iter=10000),
    'ElasticNet': ElasticNet(max_iter=10000)
}

model_types = {
    'Linear': 1,
    'Poly2': 2,
    'Poly3': 3
}

results = []

for m_type, degree in model_types.items():
    for version, base_model in model_versions.items():
        pipeline = create_pipeline(base_model, poly_degree=degree)
        
        if version != 'No_Penalization':
            param_grid = {'model__alpha': [0.01, 0.1, 1.0, 10.0]}
            if version == 'ElasticNet':
                param_grid['model__l1_ratio'] = [0.1, 0.5, 0.9]
            grid = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
            grid.fit(X, y)
            best_model = grid.best_estimator_
            best_params = grid.best_params_
        else:
            pipeline.fit(X, y)
            best_model = pipeline
            best_params = {}
        
        mse, r2 = evaluate_model(best_model, X, y, cv=5)
        results.append({
            'Model_Type': m_type,
            'Version': version,
            'Poly_Degree': degree,
            'Best_Params': best_params,
            'MSE': mse,
            'R2': r2,
            'Pipeline': best_model  # Guardamos el pipeline para referencia
        })
        
        # Para el modelo sin penalización (grado 1), mostramos el resumen de statsmodels
        if version == 'No_Penalization' and degree == 1:
            X_sm = sm.add_constant(X)
            model_sm = sm.OLS(y, X_sm).fit()
            print("=== Statsmodels Summary for Linear Regression (No Penalization) ===")
            print(model_sm.summary())

results_df = pd.DataFrame(results)
results_df_sorted = results_df.sort_values(by=['MSE', 'R2'], ascending=[True, False])
print("=== Resultados de Modelos ===")
print(results_df_sorted[['Model_Type', 'Version', 'Poly_Degree', 'Best_Params', 'MSE', 'R2']])

# --- Paso 4: Seleccionar el Modelo Óptimo y analizar significancia ---
# Supongamos que elegimos el primer modelo de la lista (óptimo en términos de MSE y R²)
best_row = results_df_sorted.iloc[0]
optimal_pipeline = best_row['Pipeline']
optimal_version = best_row['Version']
optimal_degree = best_row['Poly_Degree']
print(f"\nEl modelo óptimo seleccionado es: {best_row['Model_Type']} - {optimal_version} (Poly Degree = {optimal_degree})")

# Para modelos sin penalización, ya tenemos p-values. 
# Para modelos regularizados, extraemos los coeficientes y seleccionamos aquellos no nulos.
if optimal_version != 'No_Penalization':
    # Extraer el paso 'model' y el paso 'poly' (si existe)
    pipeline_steps = optimal_pipeline.named_steps
    if optimal_degree > 1:
        poly_transformer = pipeline_steps['poly']
        X_poly = poly_transformer.transform(X)
    else:
        X_poly = X.values  # Sin transformación polinómica
    
    scaler = pipeline_steps['scaler']
    X_scaled = scaler.transform(X_poly)
    
    # Obtener coeficientes del modelo regularizado
    coef = optimal_pipeline.named_steps['model'].coef_
    coef_df = pd.DataFrame({'Predictor': predictors, 'Coefficient': coef[:len(predictors)]})
    
    # Definir un umbral para considerar coeficientes relevantes (por ejemplo, valor absoluto > 1e-4)
    threshold = 1e-4
    significant_predictors = coef_df[np.abs(coef_df['Coefficient']) > threshold]['Predictor'].tolist()
    
    print("\nPredictores con coeficientes relevantes en el modelo regularizado:")
    print(significant_predictors)
    
    # Reajustar un modelo OLS usando solo los predictores significativos para obtener p-values
    X_selected = X[significant_predictors]
    X_sm = sm.add_constant(X_selected)
    ols_selected = sm.OLS(y, X_sm).fit()
    print("\n=== Statsmodels Summary for OLS on Selected Predictors ===")
    print(ols_selected.summary())
    
    # Visualización: Graficar los coeficientes y p-values
    summary_df = ols_selected.summary2().tables[1]
    summary_df = summary_df.drop('const')  # Excluir constante para visualizar sólo los predictores
    
    plt.figure(figsize=(10,6))
    sns.barplot(x=summary_df.index, y=summary_df['Coef.'])
    plt.errorbar(x=range(len(summary_df.index)), y=summary_df['Coef.'],
                 yerr=[summary_df['Coef.'] - summary_df['[0.025'], summary_df['[0.975]'] - summary_df['Coef.']],
                 fmt='none', c='black', capsize=5)
    plt.xticks(rotation=45)
    plt.title("Coeficientes de OLS (Predictores Seleccionados)")
    plt.ylabel("Coeficiente")
    plt.xlabel("Predictor")
    plt.tight_layout()
    plt.show()
else:
    # Si el modelo óptimo es sin penalización y grado 1, usamos directamente el modelo OLS obtenido.
    X_sm = sm.add_constant(X)
    optimal_ols = sm.OLS(y, X_sm).fit()
    print("\n=== Statsmodels Summary for Optimal OLS Model ===")
    print(optimal_ols.summary())

# --- Paso 5: Visualización Comparativa de Métricas ---
plt.figure(figsize=(14, 6))
sns.barplot(x="Model_Type", y="MSE", hue="Version", data=results_df_sorted)
plt.title("Comparación de MSE por Tipo de Modelo y Versión")
plt.xlabel("Tipo de Modelo")
plt.ylabel("MSE")
plt.legend(title="Versión")
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x="Model_Type", y="R2", hue="Version", data=results_df_sorted)
plt.title("Comparación de R² por Tipo de Modelo y Versión")
plt.xlabel("Tipo de Modelo")
plt.ylabel("R²")
plt.legend(title="Versión")
plt.show()
```

```python
poly_degree = 2
alpha_value = 10.0

poly = PolynomialFeatures(degree=poly_degree, include_bias=False)
scaler = StandardScaler()
ridge = Ridge(alpha=alpha_value, max_iter=10000)

model_pipeline = Pipeline([
    ('poly', poly),
    ('scaler', scaler),
    ('model', ridge)
])

model_pipeline.fit(X, y)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = []
r2_scores = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model_pipeline.fit(X_train, y_train)
    y_pred = model_pipeline.predict(X_test)
    
    mse_scores.append(mean_squared_error(y_test, y_pred))
    r2_scores.append(r2_score(y_test, y_pred))

mse_mean = np.mean(mse_scores)
r2_mean = np.mean(r2_scores)

print(f"=== Modelo Óptimo: Ridge (alpha={alpha_value}), Polinómico Grado = {poly_degree} ===")
print(f"MSE promedio (CV) = {mse_mean:.6f}")
print(f"R2 promedio (CV)  = {r2_mean:.6f}")

poly_features = model_pipeline.named_steps['poly']
feature_names = poly_features.get_feature_names_out(input_features=predictors)

final_ridge = model_pipeline.named_steps['model']
ridge_coefs = final_ridge.coef_

coef_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': ridge_coefs
})
coef_df['AbsCoef'] = coef_df['Coefficient'].abs()
coef_df.sort_values(by='AbsCoef', ascending=False, inplace=True)

print("\n=== Coeficientes del Modelo Ridge (Poly2) ===")
print(coef_df.head(15))

nonzero_indices = np.where(ridge_coefs != 0)[0]
selected_features = [feature_names[i] for i in nonzero_indices]
print("\nVariables con coeficiente distinto de 0 en Ridge:")
print(selected_features)

X_poly = poly.transform(X)
X_poly_scaled = scaler.transform(X_poly)
X_poly_scaled_df = pd.DataFrame(X_poly_scaled, index=X.index, columns=feature_names)

X_selected = X_poly_scaled_df[selected_features]

# Agrupar las variables en grupos de 10 y ajustar OLS por grupo

grouped_features = [selected_features[i:i+10] for i in range(0, len(selected_features), 10)]
group_results = []

print("\n=== Ajustando modelos OLS por grupos de 10 variables ===")
for group in grouped_features:
    X_group = X_selected[group]
    X_group_sm = sm.add_constant(X_group)
    try:
        model_group = sm.OLS(y, X_group_sm).fit()
        group_summary = model_group.summary2().tables[1].drop('const', errors='ignore')
        group_summary['Feature'] = group_summary.index
        group_results.append(group_summary)
        print(f"Grupo: {group} ajustado exitosamente.")
    except Exception as e:
        print("Error al ajustar el grupo:", group, "\nError:", e)

if group_results:
    all_group_results = pd.concat(group_results)
else:
    raise ValueError("No se pudo ajustar ningún modelo por grupo.")

all_group_results = all_group_results[all_group_results['P>|t|'].notna()]

all_group_results_sorted = all_group_results.sort_values(by='P>|t|')

top10 = all_group_results_sorted[all_group_results_sorted['P>|t|'] < 0.05].head(10)
if len(top10) < 10:
    top10 = all_group_results_sorted.head(10)

top10_features = top10['Feature'].tolist()

print("\n=== Top 10 Variables (por p-value) obtenidas de los grupos ===")
print(top10_features)

X_top10 = X_selected[top10_features]
X_top10_sm = sm.add_constant(X_top10)
ols_top10 = sm.OLS(y, X_top10_sm).fit()

print("\n=== OLS Final con Top 10 Variables (por p-value) ===")
print(ols_top10.summary())

final_summary = ols_top10.summary2().tables[1]
if '0.975]' not in final_summary.columns:
    print("Column names in summary table:", final_summary.columns)
    upper_bound_col = final_summary.columns[-1]
else:
    upper_bound_col = '0.975]'

xerr = (final_summary[upper_bound_col] - final_summary['Coef.']).abs()

final_summary_no_const = final_summary.drop('const', errors='ignore')
final_summary_no_const.sort_values(by='Coef.', key=abs, ascending=False, inplace=True)

plt.figure(figsize=(10, 6))
plt.barh(final_summary_no_const.index, final_summary_no_const['Coef.'], 
         xerr=xerr.loc[final_summary_no_const.index], color='skyblue')
plt.title("Coeficientes del Modelo OLS Final (Top 10 Variables)")
plt.xlabel("Coeficiente")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

A continuación, se describe la metodología y los principales hallazgos, relacionándolos con los objetivos específicos del estudio.

## 2. Metodología

### 2.1. Preparación de Datos

1. **Exploración y Limpieza**  
   Se inspeccionaron valores atípicos, se completaron o eliminaron valores faltantes y se normalizaron las variables para reducir la escala. La estandarización se realizó restando la media y dividiendo por la desviación estándar.

2. **Generación de Variables Polinómicas**  
   Para capturar **relaciones no lineales**, se incluyeron términos de segundo grado.  
   Dado un vector de características $\mathbf{x} = (x_1, x_2, \dots, x_p)$, la transformación polinómica de grado 2 genera términos de la forma:
   
   $$
   x_i^2 \quad \text{y} \quad x_i \, x_j \quad \text{para} \quad i \le j,
   $$
   
   ampliando el conjunto de predictores.

### 2.2. Modelo Polinómico Ridge

La regresión Ridge minimiza la función de costo:

$$
\min_{\boldsymbol{\beta}} \left\{ 
  \sum_{i=1}^{n} \left(y_i - \beta_0 - \sum_{j=1}^{p} x_{ij}\,\beta_j \right)^2
  \;+\; \alpha \sum_{j=1}^{p} \beta_j^2
\right\},
$$

donde:
- $y_i$ es la variable respuesta (en nuestro caso, la expresión de **NR1\_N**),
- $x_{ij}$ son los predictores (proteínas y sus interacciones),
- $\boldsymbol{\beta}$ es el vector de coeficientes,
- $\alpha$ es el parámetro de regularización que penaliza la magnitud de los coeficientes.

La penalización ayuda a **reducir el sobreajuste** y estabilizar la estimación de los coeficientes en presencia de múltiples predictores.

### 2.3. Selección y Significancia de Variables

Se extrajeron los coeficientes no nulos y se evaluó su **significancia estadística** mediante un modelo OLS, que permitió obtener los $p$-values e intervalos de confianza. Tras esta evaluación se seleccionaron las **Top 10 variables** más influyentes, que son:

1. $\text{PKCA\_N} \times \text{pNR1\_N}$

3. $\text{Bcatenin\_N}$  
4. $\text{NR2A\_N}$  
5. $\text{RSK\_N} \times \text{Bcatenin\_N}$  
6. $(\text{pERK\_N})^2$  
7. $(\text{APP\_N})^2$  
8. $\text{AMPKA\_N} \times \text{PSD95\_N}$  
9. $\text{BDNF\_N} \times \text{pNR2B\_N}$  
10. $\text{ITSN1\_N} \times \text{pERK\_N}$

### 2.4. Validación

Se empleó **validación cruzada** (k=5) para calcular métricas como el **error cuadrático medio (MSE)** y el **coeficiente de determinación ($R^2$)**. Posteriormente, se ajustó un modelo OLS final sobre las variables seleccionadas para confirmar la significancia de cada factor.

## 3. Resultados

### 3.1. Modelo Final OLS

El modelo final se expresa de la siguiente forma:

$$
\begin{aligned}
\widehat{\text{NR1\_N}} \;=\;& \beta_0 
+ \beta_1 (\text{PKCA\_N} \times \text{pNR1\_N})
+ \beta_2 (\text{APP\_N} \times \text{Bcatenin\_N}) \\
& + \beta_3 (\text{Bcatenin\_N})
+ \beta_4 (\text{NR2A\_N}) \\
& + \beta_5 (\text{RSK\_N} \times \text{Bcatenin\_N})
+ \beta_6 (\text{pERK\_N})^2 \\
& + \beta_7 (\text{APP\_N})^2
+ \beta_8 (\text{AMPKA\_N} \times \text{PSD95\_N}) \\
& + \beta_9 (\text{BDNF\_N} \times \text{pNR2B\_N})
+ \beta_{10} (\text{ITSN1\_N} \times \text{pERK\_N}).
\end{aligned}
$$

El ajuste del modelo reportó:
- **$R^2 \approx 0.957$**, lo que indica que el modelo explica aproximadamente el **95.7%** de la variabilidad de *NR1\_N*.
- Todos los coeficientes son altamente significativos ($p$-values < 0.05).
